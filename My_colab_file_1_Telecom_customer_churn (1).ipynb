SEPERATING THE DATA INTO FEATURES AND THE TARGET:

```python
features=df.drop('TotalCharges',axis=1)
target=df['TotalCharges']
features
```

```python
target
```

```python
# If your target variable contains CATEGORICAL DATA then we have to go for classification -- binary class(2) and multiclass (more than 2)
# While if your target variable contains CONTINUOUS(FLOAT) OR INTEGER DATA then we have to go for regression.
```

Scaling the data using Min-Max Scaler:

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.metrics import r2_score # why r2_score as this the default in cross_val_score 
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

```python
for i in range(0,100):
  features_train, features_test,target_train,target_test=train_test_split(features,target,test_size=0.2,random_state=i)
  lr.fit(features_train,target_train)
  pred_train=lr.predict(features_train)
  pred_test=lr.predict(features_test)
  print('At the random_state ', i ,'the training accuracy is :', (r2_score(target_train,pred_train)))
  print('At the random_state ', i ,'the testing accuracy is :' ,(r2_score(target_test,pred_test)))
```

```python
for i in range(0,100):
  features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=0.2,random_state=i)
  lr.fit(features_train,target_train)
  pred=lr.predict(features_test)
  print('At the random_state ' , i ,'the  accuracy score r2_score is :' ,(r2_score(target_test,pred))  )
```

```python
features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=0.2,random_state=12)
lr.fit(features_train,target_train)
```

```python
pred_test=lr.predict(features_test)
print(r2_score(target_test,pred_test))
```

Cross - Validation of the model

```python
Train_accuracy=r2_score(target_train,pred_train)
Test_accuracy=r2_score(target_test,pred_test)
from sklearn.model_selection import cross_val_score
```

```python
for i in range(0,10):
  cv_score=cross_val_score(lr,features,target,cv=i)
  cv_mean=cv_score.mean()
  print('At the cross fold of ', i ,'the cv score is: ' ,cv_mean ,'accurate train is: ' , Train_accuracy , 'accuarate_test is: ' , Test_accuracy )
```

```python
for i in range(2,10):
  cv_score=cross_val_score(lr,features,target,cv=i)
  cv_mean=cv_score.mean()
  print('At the cross fold of ' , i ,'the cv_score is : ' , cv_mean ,'accurate train is : ' , Train_accuracy ,'and the accurate test is : ' , Test_accuracy )
```

Since the number of folds dont have such impact on the accuracy and cv_score , so the cv=5 is selected(default)

# **Ensemble Technique**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
parameters={'criterion':['mse','mae'],'max_features':["auto","sqrt","log2"]}
gscv=GridSearchCV(rf,parameters)
gscv.fit(features_train,target_train)
```

```python
print(gscv.best_params_)
```

```python
rf=RandomForestRegressor(criterion='mse',max_features='auto')
rf.fit(features_train,target_train)
rf.score(features_train,target_train)
pred_rf=rf.predict(features_test)
print(r2_score(target_test,pred_rf))
 
```

```python
print(cross_val_score(rf,features,target,cv=5).mean())
```
